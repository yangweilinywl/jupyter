```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# plt.rcParams[u'font.sans-serif']=['SimHei'] #用来正常显示中文标签
# plt.rcParams[u'axes.unicode_minus']=False #用来正常显示负号
df = pd.read_parquet("/home/mi/workspace/data/anti_fraud/python")
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2-0.1, 1.02*height, '%s'% round(height*100,2)+"%")
        
def bar_plot(df):
    is_true = df.loc[df.isFake,:]
    is_fake = df.loc[~df.isFake,:]
    column_plot = ["eventid","dfOs","dfNetworkType","dfBrand","dfBatteryStatus","dfUseProxy","dfContainCheatingApps"]
    column_name = ["EventType","Os","NetworkType","Brand","BatteryStatus","UseProxy","ContainCheatingApps"]
    
    for index,column in enumerate(column_plot):
        data_plot_true = is_true[column].value_counts()[:10] if len(is_true[column].value_counts())>10 else is_true[column].value_counts()
        data_plot_fake = is_fake[column].value_counts()[:10] if len(is_fake[column].value_counts())>10 else is_fake[column].value_counts()
        
        name_list_true = list(data_plot_true.index)
        name_list_fake = list(data_plot_fake.index)
        
        num_list_true = np.array(data_plot_true.data)/sum(data_plot_true.data)
        num_list_fake = np.array(data_plot_fake.data)/sum(data_plot_fake.data)
        x_true = range(len(name_list_true))
        x_fake = range(len(name_list_fake))
        plt.figure(figsize=(16,9))
        plt.xlabel(column_name[index])
        plt.ylabel("Num")
        plt.title(column_name[index]+"-"+"Num")
        if len(name_list_true)>len(name_list_fake):
            rects_true = plt.bar(x_true,width=0.4,label="norm",height=num_list_true,color="red")
            rects_fake = plt.bar([i+0.4 for i in x_fake],width=0.4,label="fake",height=num_list_fake,color="blue")
            plt.xticks([i+0.2 for i in x_true],name_list_true)
        else:
            rects_true = plt.bar(x_true,width=0.4,label="norm",height=num_list_true,color="red")
            rects_fake = plt.bar([i+0.4 for i in x_fake],width=0.4,label="fake",height=num_list_fake,color="blue")
            plt.xticks([i+0.2 for i in x_fake],name_list_fake)
        autolabel(rects_true)
        autolabel(rects_fake)
        plt.legend()
        plt.savefig("/home/mi/workspace/data/anti_fraud/plot/{}".format(column_name[index]))
        plt.close()
        
def box_plot(df):
    res = df
    column_plot = ["dfUpTime","dfBatteryLevel"]
    column_name = ["ActiveTime","BatteryLevel"]
    for name in column_plot:
        res[name] = [float(i) if len(i)>0 else np.NaN for i in res[name]]
    res.dropna(inplace=True,subset=column_plot)
    is_true = res.loc[res.isFake,:][column_plot]
    is_fake = res.loc[~res.isFake,:][column_plot]
    is_true.columns = column_name
    is_fake.columns = column_name
    for col in column_name:
        plt.figure(figsize=(16,9))
        x = plt.subplot(121)
        if col=="ActiveTime":
            plt.ylim(1e7,1e9)
        axes1 = is_true[[col]].boxplot(showmeans=True,patch_artist=True)
        axes1.set_xlabel("Norm")
        axes2 = plt.subplot(122,sharey=axes1)
        axes2.set_xlabel("Fake")
        is_fake[[col]].boxplot(showmeans=True,patch_artist=True)
        plt.savefig("/home/mi/workspace/data/anti_fraud/plot/{}".format(col))
        plt.close()
    del res
    
bar_plot(df)    
box_plot(df)  
```
